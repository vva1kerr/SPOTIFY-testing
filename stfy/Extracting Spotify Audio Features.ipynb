# Part 1: Extracting Audio Features From Spotify Playlist

Spotify API allows you to access certain features from audio files, as well as a detailed analysis. In this part, I use the [Spotipy](https://spotipy.readthedocs.io/en/latest/#installation) library to extract relevant data from a playlist that I listen to. The goal is to get data that trains machine learning models to learn what kind of songs I like, and then use it to predict if I will like a certain song.

## 1. Setting Up

Spotify API requests need to be authenticated with a developer account. I load the credentials from a .json file that can be easily edited with anyone's credentials. Details on how to get credentials are on the [official documentation](https://spotipy.readthedocs.io/en/latest/#authorized-requests).

Each Spotify playlist, track, artist etc. has a unique resource identifier (URI). I load the playlist URI from another .json file that can be edited with any playlist URI. Each playlist URI is associated with a boolean that determines whether I like the songs in that playlist. There is no way to mark if I like or dislike a ceratin song on a playlist and have that information be available to anyone, so I group my likings by playlist.

The .json file containing playlists can have multiple playlists, so I set an index to 0 to get the first playlist from that file.

```python
import spotipy
from spotipy.oauth2 import SpotifyClientCredentials
import json
```

```python
credentials = json.load(open('credentials.json'))
client_id = credentials['client_id']
client_secret = credentials['client_secret']

playlist_index = 0

playlists = json.load(open('playlists.json'))
playlist_uri = playlists[playlist_index]['uri']
like = playlists[playlist_index]['like']
```

```python
client_credentials_manager = SpotifyClientCredentials(
    client_id=client_id,
    client_secret=client_secret)
sp = spotipy.Spotify(client_credentials_manager=client_credentials_manager)
```

## 2. Extracting Track Information from Playlist

The playlists URIs contain the username of the playlist's creator and the unique ID of the playlist, which are used to call the Spotipy function that returns playlist information.

```python
uri = playlist_uri    # the URI is split by ':' to get the username and playlist ID
username = uri.split(':')[2]
playlist_id = uri.split(':')[4]
```

```python
results = sp.user_playlist(username, playlist_id, 'tracks')
```

## 3. Extracting  IDs, Titles, and Artists

The playlist data is returned as a dictionary, and the 'tracks' key contains the individual track information. I store the ID, title and artist information in lists, to be added to a dataframe in the next step.

For artists, I have a list of lists, where the internal list contains all artists involved in a track. The first artists are just the first or primary artists in those lists.

```python
playlist_tracks_data = results['tracks']
playlist_tracks_id = []
playlist_tracks_titles = []
playlist_tracks_artists = []
playlist_tracks_first_artists = []

for track in playlist_tracks_data['items']:
    playlist_tracks_id.append(track['track']['id'])
    playlist_tracks_titles.append(track['track']['name'])
    # adds a list of all artists involved in the song to the list of artists for the playlist
    artist_list = []
    for artist in track['track']['artists']:
        artist_list.append(artist['name'])
    playlist_tracks_artists.append(artist_list)
    playlist_tracks_first_artists.append(artist_list[0])
```

## 4. Extracting Audio Features from Each Track

Using the IDs I extracted above, I obtain the [audio features](https://developer.spotify.com/documentation/web-api/reference/tracks/get-several-audio-features/) such as danceability, acousticness etc. for all the tracks, and store it in a Pandas dataframe.

```python
features = sp.audio_features(playlist_tracks_id)
```

```python
import numpy as np
import pandas as pd
```

```python
features_df = pd.DataFrame(data=features, columns=features[0].keys())
```

## 5. Merging Audio Features with Track Information

The audio features dataframe does not contain the title and artist information, so I merge the lists of titles and artists into the dataframe, and rearrange it in a more easily readable order.

```python
features_df['title'] = playlist_tracks_titles
features_df['first_artist'] = playlist_tracks_first_artists
features_df['all_artists'] = playlist_tracks_artists
#features_df = features_df.set_index('id')
features_df = features_df[['id', 'title', 'first_artist', 'all_artists',
                           'danceability', 'energy', 'key', 'loudness',
                           'mode', 'acousticness', 'instrumentalness',
                           'liveness', 'valence', 'tempo',
                           'duration_ms', 'time_signature']]
features_df.head()
```

## 6. Initial Data Exploration

I plot all the features to see which ones could be relevant.

```python
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
sns.pairplot(features_df)
```

There seems to be some positive linear correlation between energy and loudness, but for the most part, the correlation data is very scattered. Since I am looking for how the features correlate to my taste, the correlation between the individual features is not very important.

There seems to be traces of a normal distribution for each feature, which could say something about my taste. Since this data is from a playlist of songs that I like, there is a possibility that the centers and variability of these distributions says something about my taste in music. The only outlier here is time signature, which only has values of 4, but again, there could be songs with different time signatures that I don't like, so I keep that feature along with all the others.

It seems natural to have a bias towards certain artists, so I check a countplot of the primary artists as an overview. Turns out, I do have bias towards some artists, but for the most part, artists feature only once. That is why I decide to leave this out of my data.

If I did choose to keep artist information in my data, I would have to consider the list of artists involved in each song. I would have to create a column for every artist in the training data, with the value 1 indicating that artist is involed in the song, and 0 if not. There would be an obvious correlation with my taste, since I only listen to a small number of artists compared to all the artists on Spotify. However, if I were to make a prediction on a song that I would like, but performed by an artist that I never listened to, it is likely to give me a wrong answer if there is my taste is more strongly correlated to artist than audio features.

On a personal note, the artist is an important factor for me liking a song, the reasons ranging from their musical style to the cadence of their voice. If I were to analyze artist information, I would go by individual artist features (their voice, for example) instead of name, for a more objective analysis.

```python
features_df = features_df.drop(['first_artist', 'all_artists'], axis=1)
```

```python
features_df.head()
```

## 7. Additional Data from Audio Analysis

Spotify API also allows us to get [audio analysis](https://developer.spotify.com/documentation/web-api/reference/tracks/get-audio-analysis/) information. This includes information for the beats, bars, sections etc. in a particular song. I keep the number of bars, sections and segments. Beats and tatums are almost scaled versions of the number of bars so I leave those out.

```python
num_bars = []
num_sections = []
num_segments = []

for i in range(0,len(features_df['id'])):
    analysis = sp.audio_analysis(features_df.iloc[i]['id'])
    num_bars.append(len(analysis['bars'])) # beats/time_signature
    num_sections.append(len(analysis['sections']))
    num_segments.append(len(analysis['segments']))
    # print(len(analysis['tatums'])) # beats*2
    # print(len(analysis['beats'])) # tempo*minutes
```

```python
plt.figure(figsize=(16,4))
plt.subplot(1,3,1)
plt.hist(num_bars, bins=20)
plt.xlabel('num_bars')
plt.subplot(1,3,2)
plt.hist(num_sections, bins=20)
plt.xlabel('num_sections')
plt.subplot(1,3,3)
plt.hist(num_segments, bins=20)
plt.xlabel('num_segments')
```

Like the audio features, there are traces of a normal distribution, so I keep the data.

```python
features_df['num_bars'] = num_bars
features_df['num_sections'] = num_sections
features_df['num_segments'] = num_segments
features_df.head()
```

## 8. Classifying Tracks

I simply add a column to that classifies each song, 1 being I like the song, 0 being I don't. This is based on the boolean extracted from the .json file.

```python
if like == True:
    features_df['class'] = np.ones((len(features_df), 1), dtype=int)
else:
    features_df['class'] = np.zeros((len(features_df), 1), dtype=int)
```
